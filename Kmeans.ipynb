```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
column_names = ('X','Y')
df=pd.read_csv('dataset1.txt', sep=' ',names = column_names)
```

```python
df.head()
```

```python
plt.scatter(df["X"],df["Y"],c='black')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
kmeans = KMeans(n_clusters=4)
label = kmeans.fit_predict(df)
```

```python
print(kmeans.cluster_centers_)
```

```python
print(kmeans.labels_)
```

```python
f, (X, Y) = plt.subplots(1, 2, sharey=True,figsize=(10,6))
X.set_title('K Means')
X.set_xlabel('X')
X.set_ylabel('Y')
X.scatter(df['X'],df['Y'],c=kmeans.labels_,cmap='brg')
Y.set_title("Original")
Y.set_xlabel('X')
Y.set_ylabel('Y')
Y.scatter(df['X'],df['Y'],cmap='brg')
```

```python
kmeans = KMeans(n_jobs = -1, n_clusters = 5, init='k-means++')
kmeans.fit(df)
pred = kmeans.predict(df)
```

```python
frame = pd.DataFrame(df)
frame['cluster'] = pred
frame['cluster'].value_counts()
```
