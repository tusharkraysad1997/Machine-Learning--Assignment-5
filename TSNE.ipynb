```python
import numpy as np
from sklearn.datasets import load_digits
from sklearn.manifold import TSNE
import seaborn as sns
```

```python
df= load_digits()
```

```python
x = df.data
y = df.target
```

```python
x.shape
```

```python
y.shape
```

```python
tsne = TSNE(n_iter=250)
```

```python
XTSNE = tsne.fit_transform(x)
```

```python
XTSNE.shape
```

```python
df['tsne-2d-one'] = XTSNE[:,0]
df['tsne-2d-two'] = XTSNE[:,1]
sns.scatterplot(
    x="tsne-2d-one", y="tsne-2d-two",
    palette=sns.color_palette("hls", 10),
    hue="target",
    data=df,
    legend="full",
    alpha=0.3
)
```

```python
tsne = TSNE(n_iter=1000)
XTSNE = tsne.fit_transform(x)
df['tsne-2d-one'] = XTSNE[:,0]
df['tsne-2d-two'] = XTSNE[:,1]
sns.scatterplot(
    x="tsne-2d-one", y="tsne-2d-two",
    palette=sns.color_palette("hls", 10),
    hue="target",
    data=df,
    legend="full",
    alpha=0.3
)
```
